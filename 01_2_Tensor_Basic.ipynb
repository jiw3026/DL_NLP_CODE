### 간단한 모델

```python
# 6월 모의고사, 9월모고, 수능 

# data 
mon6, mon9, final = 85., 100., 95.

# 가설  : final = w1*mon6 + w2*mon9 + b 
# 초기 weight 설정
w1 = tf.Variable(0.1)
w2 = tf.Variable(0.2)
b = tf.Variable(-0.5)

# Cost Function
def myCost():
    # 예측값
    y_pred = w1*mon6 + w2*mon9 + b 

    # Loss (실제값 - 예측값)^2 
    loss = tf.square(final-y_pred)
    return loss 

# optimzer 
opt = tf.keras.optimizers.Adam(learning_rate = 0.1)

EPOCH = 100
for i in range(EPOCH):
    opt.minimize(myCost, var_list = [w1, w2, b])
    print(f"Epoch : {i} - Loss : {myCost().numpy()} - w1 : {w1.numpy()} - w2 : {w2.numpy()} - b : {b.numpy()}")
```

```python
import numpy as np 
import random 

a, b = 2, -5
rng = np.random.RandomState(1)
X = 10*rng.rand(50)
y = a*X + b + rng.randn(50) # 분산 추가 
```

```python
# 가중치 초기화
a = tf.Variable(random.random())
b = tf.Variable(random.random())

# cost function
def compute_loss():
    y_pred = a*X + b
    loss = tf.reduce_mean((y-y_pred)**2)
    return loss 

optim = tf.keras.optimizers.Adam(learning_rate = 0.01)

EPOCH = 1000
for i in range(EPOCH):
    # mse minimize
    optim.minimize(compute_loss, var_list = [a, b])

    if i % 100 == 99:
        print(f"Epoch : {i} - Loss : {compute_loss().numpy()} - a : {a.numpy()} - b : {b.numpy()}")
```

```python
import matplotlib.pyplot as plt 
line_x = np.arange(min(X), max(X), 0.01)
line_y = a * line_x + b

# 그래프를 그립니다. 
plt.plot(line_x, line_y, 'r-')
plt.plot(X, y, 'bo')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```
